```python
import pandas as pd
```

```python
df = pd.read_csv('CO2 Emissions_Canada.csv') # Get the dataset
```

```python
df
```

```python
cars_df = df[['Make', 'Model']]
```

```python
cars_df = cars_df.drop_duplicates()
```

```python
cars_df
```

```python
# Organize the models with their respected make
car_models = cars_df.groupby('Make')['Model'].agg(list).reset_index()
```

```python
car_models
```

```python
# download the cleaned up data for watchapp drop down selection functionality
car_models.to_json('car_models.json', orient='records')
```

```python
carbon_emission_rates = df[['Make', 'Model', 'CO2 Emissions(g/km)']]
```

```python
# This dataset will be used to look up the co2 emission rates for the selected car make and car model
carbon_emission_rates
```

```python
carbon_emission_rates.to_json('carbon_emission_rates.json', orient='records')
```
